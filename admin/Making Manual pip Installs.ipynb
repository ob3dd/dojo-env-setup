# Making Manual Pip Installs

>  Any packages installed via PIP need to be turned into separate Jupyter-notebook pip install commands with the user flag.

`f"!pip install {package}=={version} --user"`

```python
fname_mac_mchip = '../environment-ds_mac_mchip.yml'
fname_mac_intel = '../environment-ds_mac_intel.yml'
fname_windows = '../environment-ds_windows.yml'
```

```python

with open(fname_mac_mchip) as f:
    env_mac_mchip = f.read()

print(env_mac_mchip)
```

```python
split_env = env_mac_mchip.split('- pip:')
print(split_env[-1])
```

```python
split_pip = split_env[-1].split('\n')
split_pip[:10]
```

```python
# List for final commands 
requirements_list = []

for line in split_pip:
    clean_line = line.replace('- ','').strip()
    if (len(clean_line) > 0) & ('prefix' not in clean_line):
        if clean_line.startswith("#"):
            continue
        else:
            requirements_list.append(clean_line)
        


final_requirements  = sorted(list(set(requirements_list)))
final_requirements
```

```python
install_commands = []
for req in final_requirements:
    final_line = f"!pip install {req} --user"
    install_commands.append(final_line)
[print(cmd) for cmd in install_commands];
```
